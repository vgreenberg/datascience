```python
#Final Recommend Function - User Must be Input
def recommend(dataMat,SimilarityMatrix,user):
    final_ratings=[]
    SimilarityMatrix=np.array(SimilarityMatrix)
    ratedItems=np.count_nonzero(dataMat[user,:])#Computing the average user rating in order to calculate the final predicted rating below
    averageuserrating=(sum(dataMat[user,:]))/ratedItems
    unratedItems = np.nonzero(dataMat[user,:]==0)
    unratedItems=np.array(unratedItems)
    unratedItems=unratedItems.T
    similarity=SimilarityMatrix[user,:]
    distances=1-similarity
    sortedDistIndicies = distances.argsort()
    for item in unratedItems:
        i=1     #because closest index is always going to be the user, we want the index of the next closest user
        while dataMat[sortedDistIndicies[i],item].all()==0:
            i+=1 #if the closest user has not rated the current item, then the function goes through the list of users until it finds the closest user who has rated that item
        NearestRating=dataMat[sortedDistIndicies[i],item]
        RatedItemsCompUser=np.count_nonzero(dataMat[sortedDistIndicies[i],:])
        AverageCompUserRating=(sum(dataMat[sortedDistIndicies[i],:]))/RatedItemsCompUser
        similaritymeasure=SimilarityMatrix[sortedDistIndicies[i],user]
        PredictedRating=(averageuserrating)+(((NearestRating-AverageCompUserRating)*similaritymeasure)/similaritymeasure)
        PredictedRating=np.round(PredictedRating,decimals=0)
        if PredictedRating>5:
            PredictedRating=5
        if PredictedRating<1:
             PredictedRating=1
        final_ratings.append((user,item,PredictedRating))
    return final_ratings
```

```python
#Final MAE Function

def MAE(dataMat,SimilarityMatrix):
    final_ratings=[]
    SimilarityMatrix=np.array(SimilarityMatrix)
    for user in range(len(dataMat)-1):
        ratedItems=np.count_nonzero(dataMat[user,:])#Computing the average user rating in order to calculate the final predicted rating below
        averageuserrating=(sum(dataMat[user,:]))/ratedItems
        ratedItems = np.nonzero(dataMat[user,:]!=0)
        ratedItems=np.array(ratedItems)
        ratedItems=ratedItems.T
        similarity=SimilarityMatrix[user,:]
        distances=1-similarity
        sortedDistIndicies = distances.argsort()
        count=0
        for item in ratedItems:
            i=1 #because closest index is always going to be the user, we want the index of the next closest user
            while dataMat[sortedDistIndicies[i],item].all()==0:
                i+=1 #if the closest user has not rated the current item, then the function goes through the list of users until it finds the closest user who has rated that item
                if i==6040:
                    i=0
                    break
            NearestRating=dataMat[sortedDistIndicies[i],item]
            RatedItemsCompUser=np.count_nonzero(dataMat[sortedDistIndicies[i],:])
            AverageCompUserRating=(sum(dataMat[sortedDistIndicies[i],:]))/RatedItemsCompUser
            similaritymeasure=SimilarityMatrix[sortedDistIndicies[i],user]
            NonRoundedPredictedRating=(averageuserrating)+(((NearestRating-AverageCompUserRating)*similaritymeasure)/similaritymeasure)
            PredictedRating=np.round(NonRoundedPredictedRating,decimals=0)
            ActualRating=dataMat[user,item]
            error=abs(NonRoundedPredictedRating-ActualRating)
            if PredictedRating>5:
                PredictedRating=5
            if PredictedRating<1:
                PredictedRating=1        
            final_ratings.append((user,item,PredictedRating,ActualRating,error))
    totalratedItems=np.count_nonzero(dataMat[:])
    totalerror=np.array(final_ratings)
    totalerror=totalerror[:,4]
    totalerror=np.sum(totalerror)
    MAE=(totalerror/totalratedItems)
    return final_ratings,MAE
```
